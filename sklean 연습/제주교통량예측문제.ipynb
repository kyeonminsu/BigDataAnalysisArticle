```python
from sklearn.preprocessing import LabelEncoder
train_X.loc[:,['시도명','읍면동명']]=train_X.loc[:,['시도명','읍면동명']].apply(LabelEncoder().fit_transform).loc[:,['시도명','읍면동명']]=train_X.loc[:,['시도명','읍면동명']].apply(LabelEncoder().fit_transform)
test_X.loc[:,['시도명','읍면동명']]=test_X.loc[:,['시도명','읍면동명']].apply(LabelEncoder().fit_transform)
train_X[['시도명','읍면동명']].head(6)
```

```python
train_X=train_X.drop('id', axis=1)
test_X=test_X.drop('id', axis=1)
```

```python
def datePreprocessing(df2):
    
    df =df2.copy()
    df['일자'] = pd.to_datetime(df['일자'])
    df['년도'] = df['일자'].dt.year
    df['월'] = df['일자'].dt.month
    df['일'] = df['일자'].dt.day
    df['weekend'] = df['일자'].dt.weekday    
    
    
    return df

pre_train = datePreprocessing(train_X).drop(['일자'],axis=1)
pre_test = datePreprocessing(test_X).drop(['일자'],axis=1)
```

```python
def stand(df):
    return (df - df.mean(axis=0))/(df.std(axis=0))
```

```python
pre_train[['거주인구','근무인구','방문인구','총 유동인구']]=stand(pre_train[['거주인구','근무인구','방문인구','총 유동인구']])
pre_train.head()
```

```python
pre_test[['거주인구','근무인구','방문인구','총 유동인구']]=stand(pre_test[['거주인구','근무인구','방문인구','총 유동인구']])
pre_test.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42, n_estimators=100, max_depth=10)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pre_train, train_y['교통량'], test_size=0.15, random_state=42)
model.fit(X_train,y_train)
print(model.score(X_train,y_train))
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(pred,y_test)
print(mse)
```

```python
sub['교통량'] =model.predict(pre_test)
sub.to_csv('./submission.csv')
```

```python
def FinalMseScore():
    import pandas as pd
    y_true = pd.read_csv("https://raw.githubusercontent.com/Datamanim/jeju/main/Jeju_testy.csv",encoding='euc-kr')
    sub = pd.read_csv('./submission.csv',index_col='id')
    pred = sub.iloc[:,-1].values
    from sklearn.metrics import mean_squared_error
    mse = mean_squared_error(pred,y_true['교통량'])    
    print('submission mse score : ',mse)
    return mse
```
