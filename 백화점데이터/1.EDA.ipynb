# 이상치 확인

```python
#이상치 확인한다.
quantile1=df['최대구매액'].quantile(0)
quantile2=df['최대구매액'].quantile(0.25)
quantile3=df['최대구매액'].quantile(0.5)
quantile4=df['최대구매액'].quantile(0.75)
quantile5=df['최대구매액'].quantile(1)

print(quantile1)
print(quantile2)
print(quantile3)
print(quantile4)
print(quantile5)
print("---------")
IQR = quantile3 - quantile1
print(IQR)
print(quantile1 - 1.5 * IQR)
print(quantile3 + 1.5 * IQR)
print("---------")
search_df = df[(df['최대구매액'] < (quantile1 - 1.5 * IQR)) | (df['최대구매액'] > (quantile3 + 1.5 * IQR))]
outli=pd.DataFrame(search_df)
print(outli)
search_df.count()
#df=df.drop(search_df, axis=0)
```

```python
import numpy as np

def outliers_iqr(df):
    quantile1,quantile3=np.percentile(df,[25,75])
    iqr=quantile3-quantile1
    lower = quantile1-(iqr*1.5)
    upper = quantile3+(iqr*1.5)
    return np.where((df>upper)|(df<lower))

out=outliers_iqr(df['총구매액'])
print(out)
```
